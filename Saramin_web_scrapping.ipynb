```python
#library import 하기

import requests
from bs4 import BeautifulSoup
import re
from urllib.parse import urljoin
```

```python
# 클라우드 직업 스크래핑

job_list_cloud = []
page_num = 1
main_url = 'http://www.saramin.co.kr'

while True:
    url = f'http://www.saramin.co.kr/zf_user/jobs/list/job-category?page={page_num}&cat_key=41704%2C40220&search_optional_item=n&search_done=y&panel_count=y&isAjaxRequest=0&page_count=50&sort=RL&type=job-category&is_param=1&isSearchResultEmpty=1&isSectionHome=0&searchParamCount=1'
    req_header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36",
    }
    html = requests.get(url, req_header).text
    soup = BeautifulSoup(html, 'html.parser')
    
    # 만약 더이상 페이지 리스트에 공고가 없을 경우 break
    if soup.select('div.list_body div.list_item') == []:
        break
    for tag in soup.select('div.list_body div.list_item'):
        company_dict = {}
        company_dict["기업명"] = tag.select('div.company_nm a span')[0].text
        company_dict["직무"] = tag.select('div.notification_info a span')[0].text
        company_dict["경력사항"] = tag.select('div.recruit_condition p.career')[0].text
        company_dict["학력"] = tag.select('div.recruit_condition p.education')[0].text
        # 근무형태가 빈칸인 직무들도 있음
        try:
            company_dict["근무형태"] = tag.select('div.company_info p.employment_type')[0].text
        except:
            company_dict["근무형태"] = ""
        try:    
            company_dict["근무지"] = tag.select('div.company_info p.work_place')[0].text
        except:
            company_dict["근무지"] = ""
        category_list = []
        # 분류는 여러가지가 있기 때문에 리스트 형태로 저장
        for item in tag.select('div.notification_info div.job_meta span span'):
            category_list.append(item.text)
        company_dict["분류"] = str(category_list)
        company_dict["분류필터"] = "cloud"
        
        detail_url = tag.select_one('div.notification_info a').attrs['href']
        full_detail_url = urljoin(main_url, detail_url)
        company_dict["상세링크"] = full_detail_url
        job_list_cloud.append(company_dict)
    page_num += 1
job_list_cloud
    
```

```python
# Web Scrapping Function
def scrapping(url_front, url_back, name):
    page_num = 1
    job_list = []
    while True:
        url = urljoin(url_front, str(page_num), url_back)
        req_header = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.105 Safari/537.36",
        }
        html = requests.get(url, req_header).text
        soup = BeautifulSoup(html, 'html.parser')

        # 만약 더이상 페이지 리스트에 공고가 없을 경우 break
        if soup.select('div.list_body div.list_item') == []:
            break


        for tag in soup.select('div.list_body div.list_item'):

            # 기본정보
            job_dict = {}
            job_dict["기업명"] = tag.select('div.company_nm a span')[0].text
            job_dict["직무"] = tag.select('div.notification_info a span')[0].text
            job_dict["경력사항"] = tag.select('div.recruit_condition p.career')[0].text
            job_dict["학력"] = tag.select('div.recruit_condition p.education')[0].text
            job_dict["분류필터"] = name


            # 근무형태 / 근무지 - 빈칸인 경우도 있음
            try:
                job_dict["근무형태"] = tag.select('div.company_info p.employment_type')[0].text
            except:
                job_dict["근무형태"] = ""
            try:    
                job_dict["근무지"] = tag.select('div.company_info p.work_place')[0].text
            except:
                job_dict["근무지"] = ""

            # 분류는 여러가지가 있기 때문에 리스트 형태로 저장              
            category_list = []
            for item in tag.select('div.notification_info div.job_meta span span'):
                category_list.append(item.text)
            job_dict["분류"] = str(category_list)

            #상세 url
            detail_url = tag.select_one('div.notification_info a').attrs['href']
            full_detail_url = urljoin(main_url, detail_url)
            job_dict["상세링크"] = full_detail_url
            job_list.append(job_dict)
        page_num += 1
```
