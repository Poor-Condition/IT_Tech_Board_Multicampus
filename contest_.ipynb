```python
def detail_pg(main_url):
    import requests
    from bs4 import BeautifulSoup
    import os
    import re
    from urllib.parse import urljoin


    sub_url='https://www.wevity.com'
    res=requests.get(main_url)
    html=res.text
    res.status_code
    soup = BeautifulSoup(html, 'html.parser')
    
    
    view=soup.select_one("div.contest-detail div.tit-area span").text
    view=re.sub('<.*?>','',view)
    regex=re.compile(r'[조회수, \n\r\t:]')
    view=regex.sub('', view)
    # print(view)



    img_url= soup.select_one("div.img div.thumb img")
    img_link=urljoin(sub_url, img_url["src"])
    contest_name=img_url["alt"]


    contest_dict={}
    contest_dict['공모이름']=contest_name
    contest_dict['이미지']=img_link
    contest_dict['조회수']=view

    for li_tag in soup.select('ul.cd-info-list li'):

        contest_field=li_tag.select_one('span.tit').text

        contest_val=re.sub('<span.*?>.*?</span>','',str(li_tag))
        contest_val=re.sub('<.*?>','',contest_val)
        regex=re.compile(r'[\n\r\t]')
        contest_val=regex.sub('', contest_val)
        if not contest_val.startswith('$(function()'):        
            contest_dict[contest_field]=contest_val
    return contest_dict
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
# https://www.wevity.com/?c=find&s=1&gub=1&cidx=21&sp=&sw=&gbn=list&mode=
# 게임
# https://www.wevity.com/?c=find&s=1&gub=1&cidx=22&sp=&sw=&gbn=list&mode=
# 과학공학
# https://www.wevity.com/?c=find&s=1&gub=1&cidx=88&sp=&sw=&gbn=list&mode=
# 취업창업


#게임
contest_game=[]
game_dict={}
for val in ['soon','ing']:
    main_url='https://www.wevity.com/?c=find&s=1&gub=1&cidx=21&sp=&sw=&gbn=list&mode={}'.format(val)
    print(main_url)
    res=requests.get(main_url)
    html=res.text
    print(res.status_code)
    soup = BeautifulSoup(html, 'html.parser')
    for a_tag in soup.select('ul.list div.tit a'):
        link=urljoin(main_url, a_tag['href'])
    #     print(link)
        game_dict=detail_pg(link)
        game_dict['분류']='게임/소프트웨어'
        contest_game.append(game_dict)
print('게임')
print(contest_game)
print('-'*40)



# 과학공학
contest_sci=[]
sci_dict={}
for val in ['soon','ing']:
    main_url='https://www.wevity.com/?c=find&s=1&gub=1&cidx=22&sp=&sw=&gbn=list&mode={}'.format(val)
    print(main_url)
    res=requests.get(main_url)
    html=res.text
    print(res.status_code)
    soup = BeautifulSoup(html, 'html.parser')
    for a_tag in soup.select('ul.list div.tit a'):
        link=urljoin(main_url, a_tag['href'])
    #     print(link)
        sci_dict=detail_pg(link)
        sci_dict['분류']='과학/공학'
        contest_sci.append(sci_dict)
print('과학공학')
print(contest_sci)
print('-'*40)



# 취업창업
contest_job=[]
job_dict={}
for val in ['soon','ing']:
    main_url='https://www.wevity.com/?c=find&s=1&gub=1&cidx=88&sp=&sw=&gbn=list&mode={}'.format(val)
    print(main_url)
    res=requests.get(main_url)
    html=res.text
    print(res.status_code)
    soup = BeautifulSoup(html, 'html.parser')
    for a_tag in soup.select('ul.list div.tit a'):
        link=urljoin(main_url, a_tag['href'])
    #     print(link)
        job_dict=detail_pg(link)
        job_dict['분류']='취업/창업'
        contest_job.append(job_dict)
print('취업창업')
print(contest_job)
print('-'*40)
```
